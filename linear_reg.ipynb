```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r'ex1data1.txt',sep=",",header=None, names =  ["Population","Profit"])
df = pd.DataFrame(data)
```

```python
print(df.head())
print(df. tail())
print("Shape of the dataframe: " + str(df.shape))
print(df.describe())
df.plot()
```

```python
x = df['Population']
y = df ['Profit']

x = list(x)
y = list(y)
print('1st 5 values of x(feature) row vector: ', x[0:5], '; and size of x :', len(x))
print('1st 5 values of y(label) row vector: ', y[0:5], '; and size of y :', len(y))
plt.scatter(x,y)
```

```python
m = len(x) 
one_vector = []
def ones(m):
    #one_vector = []is a
    for i in range(0, m):
        one_vector.append(1)
    return one_vector
ones(m)
print('1st 5 values of one(feature) row vector:' , one_vector[0:5], '; and size of one_vector :', len(one_vector))
```

```python
X_train =  [one_vector] + [x] 
y_train = y[:]    # y_train and y are not alias

print(X_train)
print('X_train matrix size', len(X_train), '*', len(X_train[1]))
```

```python
theta = [0 , 0]
```

```python
layer_outputs = []
neuron_output = 0
for n_input, weight in zip(X_train, theta):
    neuron_output += n_input*weight
    layer_outputs.append(neuron_output)
print(layer_outputs)
```
